<h2>Appendix to Section II</h2>

Implementing the methods of Depth-first search and Breadth-first search.

```python
def depth_first(graph, source):
    return nx.dfs_preorder_nodes(
        G=graph,
        source=source
    )

def breadth_first(graph, source, target):
    return nx.bidirectional_shortest_path(
        G=graph,
        source=source,
        target=target
    )
```

```python
depth_first_args = list(depth_first(graph, 0))

plt.figure(figsize=(10, 10))
plt.title('Graph Visualization')
nx.draw_shell(graph.subgraph(depth_first_args), with_labels=True, node_color='tab:red')

print(depth_first_args)
```

```python
breadth_first_args = list(breadth_first(graph, 0, len(graph)-1))

plt.figure(figsize=(10, 10))
plt.title('Graph Visualization')
nx.draw_shell(graph.subgraph(breadth_first_args), with_labels=True, node_color='tab:red')

print(breadth_first_args)
```
